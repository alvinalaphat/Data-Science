```python
import requests
from bs4 import BeautifulSoup
import re
import json
```

```python
url = 'https://raw.githubusercontent.com/emeryberger/CSrankings/gh-pages/csrankings.csv'

page = requests.get(url)
soup = BeautifulSoup(page.content, 'html5lib')
```

```python
url1 = 'https://www.bestcollegereviews.org/top/bachelors-degrees-in-computer-science/'
page1 = requests.get(url1)
souper = BeautifulSoup(page1.content, 'html.parser')
universities = []
for checker in souper.select('p b')[:-1]:
    if checker.text[0] == '#':
        uni = (' '.join(checker.text.split(' ')[1:]))
        universities.append(uni.lower())
        
universities
```

```python
contacts = {}

for line in soup.find_all():
    professors = line.get_text()[36:].split('\n')
    for prof in professors:
        try:
            prof = prof.split(',')
            name = prof[0]
            website = prof[2]
            if prof[1].lower() in universities:
                contacts[name] = website
        except:
            break
    break
print(len(contacts))
print(contacts)
```

```python
import time
import warnings
warnings.filterwarnings("ignore")

tries = 0
no_duplicates = []
for key, value in contacts.items():
    home_page = ''
    
    while home_page == '':
        try:
            home_page = requests.get(value.replace('https','http'), verify=False)
            break
        except:
            print("Connection refused by server")
            time.sleep(5)
            tries += 1
            if tries % 3 == 0:
                break
            else:
                continue
    if home_page == '':
        continue
    else:
        soupy = BeautifulSoup(home_page.content, 'html5lib')
        email = re.findall('\b[A-Z0-9._%+-]+@[A-Z0-9.-]+\.[A-Z]{2,}\b', soupy.text)
        if len(email) == 0:
            email = [a["href"].replace('mailto:','') for a in soupy.select('a[href^="mailto:"]')]
        if len(email) != 0:
            email = email[0].strip()
            if '%' in email or '=' in email or email in no_duplicates or '#' in email or '&' in email or '?' in email or ':' in email or '/' in email:
                contacts[key] = ''
            else:
                contacts[key] = email
                no_duplicates.append(email)
                print(email)
        else:
            contacts[key] = ''
            
delete = [key for key in contacts if contacts[key] == '']
for key in delete:
    del contacts[key]
```
