```python
#matrix plots
```

```python
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
```

```python
flights.head()
```

```python
tips.head()
```

```python
#matrix form, show correlation
tips.corr()
```

```python
tc = tips.corr()
```

```python
sns.heatmap(tc)
```

```python
sns.heatmap(tc,annot=True,cmap='coolwarm')
```

```python
fc = flights.corr()
sns.heatmap(fc,cmap='coolwarm')
```

```python
flights
```

```python
fp = flights.pivot_table(index='month',columns='year',values='passengers')
```

```python
sns.heatmap(fp,cmap='magma',linecolor='white',linewidth=1)
```

```python
sns.clustermap(fp,cmap='coolwarm',standard_scale=1)
```

```python
#grids
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris['species'].unique()
```

```python
pg = sns.PairGrid(iris)
pg.map_diag(sns.distplot)
pg.map_upper(plt.scatter)
pg.map_lower(sns.kdeplot)
```

```python
tips.head()
```

```python
grid = sns.FacetGrid(data=tips,col='time',row='smoker')
grid.map(sns.distplot,'total_bill')
```

```python
grid = sns.FacetGrid(data=tips,col='time',row='smoker')
grid.map(plt.scatter,'total_bill','tip')
```

```python
#regression plots
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',markers=['o','v'])
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex',row='time',
           aspect=0.6,size=5)
```
